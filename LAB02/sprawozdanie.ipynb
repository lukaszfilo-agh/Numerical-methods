***Zadanie 1.***  
Zdefiniuj w notatniku funkcję $f(x)= x^{3} - 3x$ i sporządź jej wykres dla argumentów z przedziału:  
1. $x\in<-1,1>$  
2. $x\in<-5,5>$  
3. $x\in<0,5>$
  
***Wskazówki***
Jako argumentu funkcji można użyć numpy array.  

Do wizualizacji wyników w Pythonie używa się pakietu [Matplotlib](https://matplotlib.org/). Działanie tego pakietu prezentuje przykład:

```python
x = np.linspace(-10, 10)
# inicjalizacja wektora f(x) = sin(x)
y = np.sin(x)
plt.plot(x, y, label='sin(x)')
plt.xlabel("Os X")
plt.ylabel("Os Y")
plt.title("Wykres funkcji sin(x)")
plt.grid()
plt.legend()
plt.show()
```

Wspierając się dokumentacją [Matplotlib](https://matplotlib.org/) dodaj do wykresu etykiety osi, tytuł, grida i legendę.

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle("Wykres funkcji y = x**3 - 3*x")

# x e <-1,1>
x1 = np.linspace(-1, 1)
y1 = x1 ** 3 - 3 * x1
ax1.plot(x1, y1, label='y = x**3 - 3*x')
ax1.set(xlabel="X", ylabel="Y")
ax1.grid()
ax1.legend()

# x e <-5,5>
x2 = np.linspace(-5, 5)
y2 = x2 ** 3 - 3 * x2
ax2.plot(x2, y2, label='y = x**3 - 3*x')
ax2.set(xlabel="X", ylabel="Y")
ax2.grid()
ax2.legend()

# x e <0,5>
x3 = np.linspace(0, 5)
y3 = x3 ** 3 - 3 * x3
plt.plot(x3, y3, label='y = x**3 - 3*x')
ax3.set(xlabel="X", ylabel="Y")
ax3.grid()
ax3.legend()
plt.show()
```

***Wnioski:***
W tym zadaniu nauczyliśmy się rysować funkcje dla wybranej przez nas wielkości wektora.

***Zadanie 4.***  
Częstą praktyką szczególnie w dziedzinie analizy danych, statystyce, uczeniu maszynowym, itp. jest tworzenie na podstawie biblioteki [Matplotlib](https://matplotlib.org/) własnych szablonów wykresów stworzonych na podstawie funkcji. Najprostszy przykład to przeładowanie funkcji plot tak by wykres miał czerwony kolor:

```python
def my_plot(x, y, xlabel, ylabel, title, label):
    if x.shape != y.shape or min(x.shape) == 0:
        return None
    fig, ax = plt.subplots()
    ax.plot(x, y, 'r', label=label)
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
    ax.legend()
    return fig


x = np.linspace(-10, 10)
# inicjalizacja wektora f(x) = sin(x)
y = np.sin(x)
my_plot(x, y, 'x', 'y', 'Funkcja sin', 'sin(x)')
```

***Zadanie 7.***  
Za pomocą funkcji *parallel\_plot*  i przedstaw na jednym z nich [Spirale logarytmiczną](https://pl.wikipedia.org/wiki/Spirala_logarytmiczna) w szerokim przedziale, a w drugim w okolicy zera.

```python
a = 2
b = 0.2
fi1 = np.linspace(-100, 10, 100000)
fi2 = np.linspace(-100, -50, 100000)
x1 = np.array([a * np.exp(b * i) * np.cos(i) for i in fi1])
y1 = np.array([a * np.exp(b * i) * np.sin(i) for i in fi1])
x2 = np.array([a * np.exp(b * i) * np.cos(i) for i in fi2])
y2 = np.array([a * np.exp(b * i) * np.sin(i) for i in fi2])
main.parallel_plot(x1, y1, x2, y2, 'X', 'Y - duży przedział', 'X', 'Y - mały przedział', 'Spirala logarytmiczna', '|')
plt.show()
```

W tym zadaniu wykorzystaliśmy napisaną wcześniej funckje parallel_plot, która umożliwiła nam narysowanie dwóch funkcji obok siebie.

**Zadanie 10.**  
Wykorzystując funkcje *semilogx* narysuj funkcję $cos(2\pi x)$ w zakresie co najmniej od $-10^2$ do $10^0$. Uzyj do tego trzech róznych wartości kroków, gdzie jeden będzie równy około 0.02, a reszta nieznacznie większa, dla generowania wartości funkcji (za pomocą *np.arrange*), a następnie narysuj wykresy według schematu: węzły (punkty, w których znamy wartośc funckji) zaznacz pomoca kropek, połączonych liniami przerywanymi. Zauwaz jak wykresy zachowują się w miarę zwiększania kroku.  Dla porównania narysuj takze wykres niestosujac skali logarytmicznej. Wykres umieść w siatce 2x2 wykorzystując *subplot*. Nie zapomnij o oznaczeniu tytułów osi i wykresów.

```python
x1 = np.arange(-100, 1, 0.02)
x2 = np.arange(-100, 1, 0.05)
x3 = np.arange(-100, 1, 0.08)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
y3 = np.cos(2 * np.pi * x3)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 12))
fig.suptitle('Wykresy funkcji cos(2pi*x) dla różnych kroków wektora x.')

ax1.semilogx(x1, y1, 'o--', label='f(x)=cos(2pi*x)')
ax1.set(xlabel='X - skala logarytmiczna', ylabel='Y', title='Krok 0.02')
ax1.legend(loc='upper left')

ax2.semilogx(x2, y2, 'o--', label='f(x)=cos(2pi*x)')
ax2.set(xlabel='X - skala logarytmiczna', ylabel='Y', title='Krok 0.05')
ax2.legend(loc='upper left')

ax3.semilogx(x3, y3, 'o--', label='f(x)=cos(2pi*x)')
ax3.set(xlabel='X - skala logarytmiczna', ylabel='Y', title='Krok 0.08')
ax3.legend(loc='upper left')

ax4.plot(x3, y3, 'o--', label='f(x)=cos(2pi*x)')
ax4.set(xlabel='X - skala normalna', ylabel='Y', title='Krok 0.08')
ax4.legend(loc='upper left')
```

***Wnioski:***
W tym zadaniu przedstawiliśmy wykresy funckji cos(2pix) z różnymi krokami i w różnych skalach. Wykresy te znacząco różnią się od siebie w zależności od przyjętego kroku, natomiast wykres w skali normalnej jest nieczytelny i trudno jest cokolwiek z niego wywnioskować.
