```python
import main
import numpy as np
import scipy as sp
from scipy import linalg
import matplotlib
import matplotlib.pyplot as plt
```

***Zadanie 1***   

Wykorzystując funkcję *random_matrix_by_egval* wygenerować losowe macierze dla zadanych zestawów wartości własnych:

$\lambda_{1} = \{ 0 , 1 , 2 , ... , 20 \}$

$\lambda_{2} = \{ 10^{-10} , 10^{-9} , ... , 10^{-1} , 1 \}$

$\lambda_{3} = \{ j , -j , 1+j , 1-j \}$

Następnie dla wygenerowanych macierzy oblicz wartości własne za pomocą funkcji *eigvals* i wyrysuj je na wykresie.

```python
v1 = np.arange(0, 21, 1)
v2 = np.array([10 ** x for x in range(-10, 1)])
v3 = np.array([1j, -1j, 1+1j, 1-1j])
A1 = main.random_matrix_by_egval(v1)
A2 = main.random_matrix_by_egval(v2)
A3 = main.random_matrix_by_egval(v3)

A1_eig_vals = sp.linalg.eigvals(A1)
A2_eig_vals = sp.linalg.eigvals(A2)
A3_eig_vals = sp.linalg.eigvals(A3)

fig1, (a1, a2, a3) = plt.subplots(1, 3, figsize=(21, 7))
a1.scatter(A1_eig_vals.real, A1_eig_vals.imag)
a1.scatter(v1.real, v1.imag, marker='x')
a1.set(xlabel='Re', ylabel='Im')
a1.set(xticks=v1[0:11]*2)
a1.set_title('Wartości własne macierzy A1')

a2.scatter(A2_eig_vals.real, A2_eig_vals.imag)
a2.scatter(v2.real, v2.imag, marker='x')
a2.set(xlabel='Re', ylabel='Im')
a2.set_title('Wartości własne macierzy A2')

a3.scatter(A3_eig_vals.real, A3_eig_vals.imag)
a3.scatter(v3.real, v3.imag, marker='x')
a3.set(xlabel='Re', ylabel='Im')
a3.set_title('Wartości własne macierzy A3');
```

***Wnioski:***

Dla wygenerowanych macierzy wartości obliczone pokrywają się z zadanymi, zarówno dla wartości rzeczywistych jak i zespolonych.

***Zadanie 2***   

Wykorzystując zestawy wartości własnych z zadania pierwszego wygeneruj macierze funkcją *random_matrix_by_egval* (lub uzyj wygenerowanych wcześniej macierzy). 

Następnie do otrzymanych wyników dodaj macierz $B$, gdzie $B = \alpha I$, przy czym $\alpha$ równa się ilości liter w Twoim nazwisku.

Wypisz wartości własne otrzymanej w ten sposób macierzy. Porównaj je z pierwotnie wykorzystanymi.

```python
def add_B(m):
    return m + 4 * np.eye(m.shape[0])


A11 = add_B(A1)
A22 = add_B(A2)
A33 = add_B(A3)

A11_eig_vals = sp.linalg.eigvals(A11)
A11_eig_vals = np.sort(A11_eig_vals)
A22_eig_vals = sp.linalg.eigvals(A22)
A22_eig_vals = np.sort(A22_eig_vals)
A33_eig_vals = sp.linalg.eigvals(A33)
A33_eig_vals = np.sort(A33_eig_vals)

print('Wartości własne macierzy A1 * 4I: \n', A11_eig_vals.real)
print('Pierwotne wartości własne macierzy A1: \n', v1)
print('Różnica pomiędzy wartościami własnymi macierzy A1 * 4I i A1 : \n', A11_eig_vals.real - v1)
print('Wartości własne macierzy A2 * 4I: \n', A22_eig_vals.real)
print('Pierwotne wartości własne macierzy A2: \n', v2)
print('Różnica pomiędzy wartościami własnymi macierzy A2 * 4I i A2 : \n', A22_eig_vals.real - v2)
print('Wartości własne macierzy A3 * 4I: \n', A33_eig_vals)
print('Pierwotne wartości własne macierzy A3: \n', v3)
print('Różnica pomiędzy wartościami własnymi macierzy A3 * 4I i A3 : \n', A33_eig_vals - v3)
```

***Wnioski:***

Wartości własne po dodaniu macierzy jednostkowej pomnożonej przez 4 zwiększyły się dokładnie o 4.

3) Zaimplementuj funkcję przekształcająca podane wartości własne, na losową macierz (wykorzystaj zaimplementowana wczesniej funkcje *random_matrix_by_egval*). Potem oblicz wartości własne ostatecznej macierzy i wyrysuj jest na nowym wykresie (w postaci kółek).

4) Kolejno powtórz czynności punktu trzeciego, dodając zaburzenia do wygenerowanej (tej samej!) macierzy poprzez dodanie do niej macierzy $R$, gdzie kazda z jej wartości jest postaci $(10^{-10})N(0,1)$. Oblicz wartości własnej nowej macierzy i narysuj je na tym samym wykresie co uprzednie (w postaci krzyzyków)

```python
eig_vals = np.arange(0, 21, 1)
polly = np.poly(eig_vals)
frob = main.frob_a(polly[1:])
frob_eig_vals = sp.linalg.eigvals(frob)

polly_err = polly + 10 ** -10 * np.random.normal(0, 1)
frob_err = main.frob_a(polly_err[1:])
frob_eig_vals_err = sp.linalg.eigvals(frob_err)

fig2, af = plt.subplots(1, 1, figsize=(7,7))
af.scatter(frob_eig_vals.real, frob_eig_vals.imag)
af.scatter(frob_eig_vals_err.real, frob_eig_vals_err.imag, marker='x')
af.set(xlabel='Re', ylabel='Im')
af.grid()
af.set_title('Porównanie wartości niezaburzonych z zaburzonym wielomianem charakterystycznym');
```

```python
matrix = main.random_matrix_by_egval(eig_vals)
eig_matrix = np.linalg.eigvals(matrix)

new_matrix = matrix + 10 ** -10 * np.random.normal(0, 1)
new_eig_matrix = np.linalg.eigvals(new_matrix)

fig4, v = plt.subplots(figsize=(7,7))
v.scatter(eig_matrix.real, eig_matrix.imag)
v.scatter(new_eig_matrix.real, new_eig_matrix.imag, marker='x')
v.set(xlabel='Re', ylabel='Im')
v.grid()
v.set_title('Porównanie wartości niezaburzonych z zaburzoną macierzą');
```

***Wnioski:***

Zaburzenie wielomianu charakterystycznego dużo mocniej wpływa na obliczone wartości własne macierzy. Można zatem stwierdzić że wielomian charakterystyczny jest dużo bardziej podatyny na zaburzenia niż losowa macierz.

***Zadanie 4***   

Dokonaj analizy dla zestawu wartości własnych: $\lambda = \{ 1 - 10^{-K}, K = 0,1,2,...,20 \}$. na dwa sposoby:

1) Poprzez utworzenie losowej macierzy, obliczenie wartości własnych, a następnie ich wyrysowania.
2) Poprzez obliczenie pierwiatków wielomianu charakterystycznego, a następnie ich wyrysowania.

Wykorzystaj wcześniej zaimplementowane funkcje.

```python
lambda1 = [1 - 10 ** -x for x in range(21)]
matrix4 = main.random_matrix_by_egval(lambda1)
eig_vals41 = np.linalg.eigvals(matrix4)

polly4 = np.poly(lambda1)
frob4 = main.frob_a(polly4[1:])
eig_vals42 = np.linalg.eigvals(frob4)

fig5, vv = plt.subplots(figsize=(7,7))
vv.scatter(eig_vals41.real, eig_vals41.imag, label='Wartości własne wyegenerowanej macierzy')
vv.scatter(eig_vals42.real, eig_vals42.imag, marker='x', label='Wartości własne macierzy frobeniusa')
vv.set(xlabel='Real', ylabel='Imag')
vv.grid()
vv.legend()
vv.set_title('Porównanie wartości własnych wyliczonych różnymi metodami');
```

***Wnioski:***

Wartości własne obliczone z wygenerowanej losowo macierzy w znacznym stopniu zgadzają sie z wartościami zadanymi, a wartości własne obliczone za pomocą macierzy frobeniusa w dużym stopniu odbiegają od tych zadanych. Można z tego wywnioskować że obliczanie wielomianu charaterystycznego znając jego miejsca zerowe generuje duże błędy obliczeniowe.
